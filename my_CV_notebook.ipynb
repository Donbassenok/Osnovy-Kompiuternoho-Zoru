```python
import cv2 #opencv
import numpy as np # matrix manipulations

%matplotlib inline 
from matplotlib import pyplot as plt 
```

## Image load

```python
input_image = cv2.imread('photo_2025-10-03_20-17-29.jpg')
b,g,r = cv2.split(input_image)
```

## Image show channels

```python
plt.imshow(r, cmap='gray')
```

## Image color

```python
merged_img = cv2.merge([r,g,b])
plt.imshow(merged_img)
```

## Image crop

```python
crop_image = merged_img[400:700, 0:300, :]
plt.imshow(crop_image)
```
